```python
#pip install umap-learn
import umap

# Step 1: Generate synthetic data for two conditions
np.random.seed(42)
data_condition1 = np.random.normal(0, 1, (100, 10))  # 100 samples, 10 features
data_condition2 = data_condition1 + np.random.normal(0, 0.1, (100, 10))  # Slightly shifted

# Combine data and add labels
data = [data_condition1, data_condition2]
labels = [np.array([0] * 100), np.array([1] * 100)]  # Labels for visualization

# Step 2: Perform Aligned-UMAP
aligned_mapper = umap.AlignedUMAP(n_neighbors=15, min_dist=0.1, metric='euclidean')
aligned_embeddings = aligned_mapper.fit_transform(data)

# Step 3: Visualize results
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
for i, embedding in enumerate(aligned_embeddings):
    scatter = ax[i].scatter(embedding[:, 0], embedding[:, 1], c=labels[i], cmap='viridis', s=15)
    ax[i].set_title(f'Condition {i + 1}')
    ax[i].set_xlabel('UMAP 1')
    ax[i].set_ylabel('UMAP 2')
plt.colorbar(scatter, ax=ax, orientation='vertical', label='Labels')
plt.tight_layout()
plt.show()
```
