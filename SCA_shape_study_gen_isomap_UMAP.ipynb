```python
##################  distance matrix filtering, ATRIL/BIOSCA/CERMOI  ###################
ATRIL_fileName = rf'{curRoot}:\B_projWIP\proj_ataxia\CSSyl_combined_ATRIL_time12_names.csv'
BIOSCA_fileName = rf'{curRoot}:\B_projWIP\proj_ataxia\CSSyl_combined_BIOSCA_time12_names.csv'
CERMOI_fileName = rf'{curRoot}:\B_projWIP\proj_ataxia\CSSyl_combined_CERMOI_time12_names.csv'

ATRIL_subjNames = pd.read_csv(ATRIL_fileName, header=0).squeeze("columns").tolist() # read in as alist, not df
BIOSCA_subjNames = pd.read_csv(BIOSCA_fileName, header=0).squeeze("columns").tolist()
CERMOI_subjNames = pd.read_csv(CERMOI_fileName, header=0).squeeze("columns").tolist()

def findOverlapDiff(subjList, distance):
    # Clean the input list and index
    clean_list = [str(s).strip() for s in subjList]
    clean_index = [str(s).strip() for s in distance.index]
    clean_col = [str(s).strip() for s in distance.columns]
    
    # Find overlapping subjects
    list_overlap = [s for s in clean_list if s in clean_index and s in clean_col]

    # Find missing subjects
    list_missing = [s for s in clean_list if s not in clean_index]
    print("Number of overlapping subjects:", len(list_overlap))
    print("Number of missing subjects:", len(list_missing))
    return list_overlap, list_missing
    
#res_lists = findOverlapDiff(BIOSCA_time12_subjNames,minDist)  # use BIOSCA_time12_subjNames, composed above in code
print("BIOSCA")
res_lists_BIOSCA = findOverlapDiff(BIOSCA_subjNames,minDist)   # use BIOSCA_subjNames, read from file, same as that composed above in code
overlap_BIOSCA = res_lists_BIOSCA[0]
missing_BIOSCA = res_lists_BIOSCA[1]
minDist_BIOSCA = minDist.loc[overlap_BIOSCA,overlap_BIOSCA]
maxDist_BIOSCA = maxDist.loc[overlap_BIOSCA,overlap_BIOSCA]
print("ATRIL")
res_lists_ATRIL = findOverlapDiff(ATRIL_subjNames,minDist)   # use ATRIL_subjNames, read from file, same as that composed above in code
overlap_ATRIL = res_lists_ATRIL[0]
missing_ATRIL = res_lists_ATRIL[1]
minDist_ATRIL = minDist.loc[overlap_ATRIL,overlap_ATRIL]
maxDist_ATRIL = maxDist.loc[overlap_ATRIL,overlap_ATRIL]
print("CERMOI")
res_lists_CERMOI = findOverlapDiff(CERMOI_subjNames,minDist)   # use CERMOI_subjNames, read from file, same as that composed above in code
overlap_CERMOI = res_lists_CERMOI[0]
missing_CERMOI = res_lists_CERMOI[1]
minDist_CERMOI = minDist.loc[overlap_CERMOI,overlap_CERMOI]
maxDist_CERMOI = maxDist.loc[overlap_CERMOI,overlap_CERMOI]
```
