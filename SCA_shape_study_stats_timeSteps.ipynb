```python
#######################  Rank the changes from one time point to another  ########################

# Note that for ctl_sca1 and sca7, sca1/7 has time points 1 and 2, ctl may have time points 1, 2 and 3, 
# since controls are partly from BIOSCA (time 1 and 2), partly CERMOI (time 1 and 3)
# combined_only_SCA for ctl_SCA1, ctl_SCA3 has only time1 and time2
# combined_only_ctl for ctl_SCA1, ctl_SCA3 has time1, time2 and time3

###  Find the ranking of subjects according to the magnitude of change, only SCA, time 1 and 2
valueSelect = "iso1"   ####  change  ####
df = combined_only_SCA ####  change  ####
#df = combined_only_ctl_BIOSCA      # selection of controls ONLY from ctl_SCA1 and ctl_SCA3, since only Biosca has SCA1 and SCA3
# df = combined_only_ctl_CERMOI    # selection of controls from CERMOI

# Pivot the dataframe so each time_point becomes a column
pivot_df = df.pivot(index="ori_subjName", columns="Time_point", values=valueSelect)
# Rename columns for better readability
pivot_df.columns = ["time_point_1", "time_point_2"]
# Calculate the change from time_point 1 to 2
pivot_df["change_1_to_2"] = pivot_df["time_point_2"] - pivot_df["time_point_1"]

# Sort based on the absolute value of the change in descending order
sorted_df = pivot_df.reindex(
    pivot_df["change_1_to_2"].abs().sort_values(ascending=False).index
)

# Add the rank based on the sorted order
sorted_df["rank"] = range(1, len(sorted_df) + 1)

# Print the result
print(sorted_df)
```

```python
################  Normality tests  ################

measures = [
    'iso1', 'iso2', 'iso3', 'UMAP1_U1', 'UMAP1_U2', 'UMAP1_U3', 'UMAP2_U3', 'UMAP1_U4', 'UMAP2_U4',
    'opening','maxdepth_talairach','iso2_asy','UMAP1_U1_asy','UMAP1_U2_asy','UMAP1_U3_asy','UMAP1_U4_asy','UMAP2_U4_asy'
]

# Shapiro-Wilk test for normality, not normally distributed if p-val < 0.05
def run_shapiro_tests(df, measure_list):
    for measure in measure_list:
        if measure in df.columns:
            stat, p_value = shapiro(df[measure].dropna())  # dropna to avoid issues with NaN
            print(f'{measure} p-val: {p_value}')
        else:
            print(f"Warning: '{measure}' not found in DataFrame columns.")

# Run the function
print('Normality test stats for time1:')
print()
run_shapiro_tests(val_time1, measures)
print()
print('Normality test stats for time2:')
print()
run_shapiro_tests(val_time2, measures)
```

```python
# test difference ctl+sca: time1 and time2, all 

######  For two groups only, no covariates  ######
####  Perform t-test/Wilcoxon rank-sum analysis for Gaissian/Non-Gaussian distributions  ###
from scipy.stats import ttest_ind     # for t-test
from scipy.stats import mannwhitneyu  # for Mann-Whitney U test (Wilcoxon rank-sum test)
measures_complete = [
    'iso1', 'iso2', 'iso3', 'UMAP1_U1', 'UMAP1_U2', 'UMAP1_U3', 'UMAP2_U3', 'UMAP1_U4', 'UMAP2_U4',
    'surface_talairach','surface_native','maxdepth_talairach','maxdepth_native','meandepth_talairach',
    'meandepth_native','hull_junction_length_talairach','hull_junction_length_native','GM_thickness','opening',
    'iso1_asy','iso2_asy','iso3_asy','UMAP1_U1_asy','UMAP1_U2_asy','UMAP1_U3_asy','UMAP2_U3_asy','UMAP1_U4_asy','UMAP2_U4_asy',
    'surface_talairach_asy','surface_native_asy','maxdepth_talairach_asy','maxdepth_native_asy','meandepth_talairach_asy',
    'meandepth_native_asy','hull_junction_length_talairach_asy','hull_junction_length_native_asy','GM_thickness_asy','opening_asy'
]
measures = [
    'iso1', 'iso2', 'iso3', 'UMAP1_U1', 'UMAP1_U2', 'UMAP1_U3', 'UMAP2_U3', 'UMAP1_U4', 'UMAP2_U4','SARA'
]

def t_Wilcoxonranksum(measures,df):  # CHANGE passing measures or measures_complete
    for measure in measures:
        print(measure)
        t_test_result = ttest_ind(df[measure][df['Time_point'] == 1], df[measure][df['Time_point'] != 1], equal_var=True)
        print(f"T-Test Result: t-statistic={t_test_result.statistic}, p-value={t_test_result.pvalue}")
        mann_whitney_result = mannwhitneyu(df[measure][df['Time_point'] == 1], df[measure][df['Time_point'] != 1])
        print(f"Mann-Whitney U Test: U-statistic={mann_whitney_result.statistic}, p-value={mann_whitney_result.pvalue}")

t_Wilcoxonranksum(measures, df=combined)

# left hemisphere, right hemisphere
print()
print('Left hemisphere')
t_Wilcoxonranksum(measures, df=combined_L)
print()
print('Right hemisphere')
t_Wilcoxonranksum(measures, df=combined_R)

# only ctl, only sca
print()
print('Only controls:')
t_Wilcoxonranksum(measures, df=combined_only_ctl)
print()
print('Only SCAs:')
t_Wilcoxonranksum(measures, df=combined_only_SCA)
```

```python
## Statistics for the difference between two time points ##
#If Age:C(TimePoint) is significant(p < 0.05) → The slopes of the regression lines are significantly different.
#If C(TimePoint) is significant → The overall mean difference between Time 1 and Time 2 is significant.
#If Age is significant → The measure changes significantly with age.

curVal = 'iso1'  # Change if needed
curX = 'Age' # Age, SARA
plot_x_1 = curX + '_time1'
plot_x_2 = curX + '_time2'
plot_val_1 = curVal + '_time1'
plot_val_2 = curVal + '_time2'

# Reshape data to long format
merged_long = merged.melt(id_vars=['ori_subjName', 'Age_time1', 'Age_time2'], 
                          value_vars=[plot_val_1, plot_val_2], 
                          var_name='TimePoint', value_name='Measure')

# Convert TimePoint to categorical (1 for time1, 2 for time2)
merged_long['TimePoint'] = merged_long['TimePoint'].map({plot_val_1: 1, plot_val_2: 2})

# Use Age corresponding to each TimePoint
merged_long['Age'] = merged_long.apply(lambda row: row['Age_time1'] if row['TimePoint'] == 1 else row['Age_time2'], axis=1)

# Fit ANCOVA model with interaction
model = smf.ols('Measure ~ Age * C(TimePoint)', data=merged_long).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```
