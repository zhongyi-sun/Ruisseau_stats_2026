```python
# selection of distances

#ctl_1_7_subj = combined[('SCA'==0) | ('SCA'==1) | ('SCA'==7)].index
ctl_1_7_subj = combined[combined['SCA'].isin([0, 1, 7])].index
ctl_1_7_min_dist = minDist.loc[ctl_1_7_subj,ctl_1_7_subj]
ctl_1_7_max_dist = maxDist.loc[ctl_1_7_subj,ctl_1_7_subj]

sca_1_7_subj = combined[combined['SCA'].isin([1, 7])].index
sca_1_7_min_dist = minDist.loc[sca_1_7_subj,sca_1_7_subj]
sca_1_7_max_dist = maxDist.loc[sca_1_7_subj,sca_1_7_subj]
```

############################  generation of Isomap  #############################

```python

from sklearn.manifold import Isomap
numDim = 3
numNeig = 10

subjNames = sca_1_7_min_dist.index
dimNames = np.arange(1,numDim+1)
iso_sca_1_7_min = Isomap(n_neighbors=numNeig,n_components=numDim,metric='precomputed').fit_transform(sca_1_7_min_dist.values)
iso_sca_1_7_min_DF = pd.DataFrame(iso_sca_1_7_min,index=subjNames,columns=dimNames)
iso_sca_1_7_max = Isomap(n_neighbors=numNeig,n_components=numDim,metric='precomputed').fit_transform(sca_1_7_max_dist.values)
iso_sca_1_7_max_DF = pd.DataFrame(iso_sca_1_7_max,index=subjNames,columns=dimNames)

subjNames = ctl_1_7_min_dist.index
dimNames = np.arange(1,numDim+1)
iso_ctl_1_7_min = Isomap(n_neighbors=numNeig,n_components=numDim,metric='precomputed').fit_transform(ctl_1_7_min_dist.values)
iso_ctl_1_7_min_DF = pd.DataFrame(iso_ctl_1_7_min,index=subjNames,columns=dimNames)
iso_ctl_1_7_max = Isomap(n_neighbors=numNeig,n_components=numDim,metric='precomputed').fit_transform(ctl_1_7_max_dist.values)
iso_ctl_1_7_max_DF = pd.DataFrame(iso_ctl_1_7_max,index=subjNames,columns=dimNames)
```

```python
# Save sca_1_7 Isomaps as csv
selectType = 'sca_1_7'
outNameMin = 'isomapCmds'+curRegion+'k'+str(numNeig)+'d'+str(numDim)+'distmin'+'_'+selectType+'.txt'
outFileNameMin = rf'D:\B_projWIP\proj_ataxia\atrilBioscaCermoi\{curRegion}\Isomap_select\{outNameMin}'
iso_sca_1_7_min_DF.to_csv(outFileNameMin,index_label='subjName')
print(outFileNameMin)

outNameMax = 'isomapCmds'+curRegion+'k'+str(numNeig)+'d'+str(numDim)+'distmax'+'_'+selectType+'.txt'
outFileNameMax = rf'D:\B_projWIP\proj_ataxia\atrilBioscaCermoi\{curRegion}\Isomap_select\{outNameMax}'
iso_sca_1_7_max_DF.to_csv(outFileNameMax,index_label='subjName')
print(outFileNameMax)
```

```python
# Save ctl_1_7 Isomaps as csv
selectType = 'ctl_1_7'
outNameMin = 'isomapCmds'+curRegion+'k'+str(numNeig)+'d'+str(numDim)+'distmin'+'_'+selectType+'.txt'
outFileNameMin = rf'D:\B_projWIP\proj_ataxia\atrilBioscaCermoi\{curRegion}\Isomap_select\{outNameMin}'
iso_ctl_1_7_min_DF.to_csv(outFileNameMin,index_label='subjName')
print(outFileNameMin)

outNameMax = 'isomapCmds'+curRegion+'k'+str(numNeig)+'d'+str(numDim)+'distmax'+'_'+selectType+'.txt'
outFileNameMax = rf'D:\B_projWIP\proj_ataxia\atrilBioscaCermoi\{curRegion}\Isomap_select\{outNameMax}'
iso_ctl_1_7_max_DF.to_csv(outFileNameMax,index_label='subjName')
print(outFileNameMax)
```

```python
# selection of Isomap by hemisphere
iso_sca_1_7_min_R_DF = iso_sca_1_7_min_DF[iso_sca_1_7_min_DF.index.str.startswith('flip')]
iso_ctl_1_7_min_R_DF = iso_ctl_1_7_min_DF[iso_ctl_1_7_min_DF.index.str.startswith('flip')]

#print(iso_ctl_1_7_min_R_DF.index)

# Save selected hemisphere Isomaps as csv
outName = 'isomapCmds'+curRegion+'k'+str(numNeig)+'d'+str(numDim)+'distmin_sca_1_7_R.txt'
outFileName = rf'D:\B_projWIP\proj_ataxia\atrilBioscaCermoi\{curRegion}\Isomap_select\{outName}'
iso_sca_1_7_min_R_DF.to_csv(outFileName,index_label='subjName')
print(outFileName)

outName = 'isomapCmds'+curRegion+'k'+str(numNeig)+'d'+str(numDim)+'distmin_ctl_1_7_R.txt'
outFileName = rf'D:\B_projWIP\proj_ataxia\atrilBioscaCermoi\{curRegion}\Isomap_select\{outName}'
iso_ctl_1_7_min_R_DF.to_csv(outFileName,index_label='subjName')
print(outFileName)
```
