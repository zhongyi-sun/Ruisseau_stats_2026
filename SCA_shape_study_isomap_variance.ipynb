```python
from sklearn.manifold import Isomap
import numpy as np

# Assuming dist_matrix is your N x N distance matrix
n_neighbors = 5  # 5 or 30
n_components = 10  # Number of dimensions in Isomap

# Apply Isomap algorithm
isomap = Isomap(n_neighbors=n_neighbors, n_components=n_components, metric='precomputed')
embedding = isomap.fit_transform(minDist)

# Calculate the variance explained by each Isomap dimension
#variances = np.var(embedding, axis=0)
#explained_variance_ratio = variances / np.sum(variances)

# Display explained variance ratio for each dimension
#for i, ratio in enumerate(explained_variance_ratio):
#    print(f"Dimension {i+1} explained variance ratio: {ratio:.4f}")

#  Access the geodesic distance matrix and compute eigenvalues
# `dist_matrix_` stores the geodesic distance matrix between points
geodesic_dist_matrix = isomap.dist_matrix_

# Perform eigenvalue decomposition on the geodesic distance matrix
# First, center the distance matrix using double centering formula
n = geodesic_dist_matrix.shape[0]
H = np.eye(n) - np.ones((n, n)) / n
K = -0.5 * H @ geodesic_dist_matrix @ H

# Eigenvalue decomposition to extract the eigenvalues
eigenvalues, eigenvectors = np.linalg.eigh(K)

# Sort eigenvalues in descending order
sorted_eigenvalues = np.flip(np.sort(eigenvalues))

# Calculate the explained variance ratio for each dimension
explained_variance_ratio = sorted_eigenvalues[:n_components] / np.sum(sorted_eigenvalues)

# Print the explained variance for each dimension
for i, ev in enumerate(explained_variance_ratio):
    print(f"Dimension {i+1} explains {ev * 100:.2f}% of the variability")
    
```

```python
import umap
import numpy as np

n_components = 2  # Number of UMAP dimensions

# Apply UMAP algorithm
umap_model = umap.UMAP(n_components=n_components, random_state=42)
embedding = umap_model.fit_transform(maxDist)

# Calculate the variance explained by each UMAP dimension
variances = np.var(embedding, axis=0)
explained_variance_ratio = variances / np.sum(variances)

# Display explained variance ratio for each dimension
for i, ratio in enumerate(explained_variance_ratio):
    print(f"Dimension {i+1} explained variance ratio: {ratio:.4f}")
```
